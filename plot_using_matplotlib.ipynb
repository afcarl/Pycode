```python
x = np.linspace(0,2*3.14,100,True)
y = sin(x)
z = cos(x)
plt.plot(y)
plt.plot(z) # Two functions in same plot
```

```python
plt.subplot(2,1,1)
plt.plot(y)
plt.subplot(2,1,2)
plt.plot(z)
plt.ylabel('sin(x)')
plt.show() #Two functions as subplot
```
